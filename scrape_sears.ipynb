```python
sel = 'div.card-container'
```

```python
prev_items_found = -1
n_items_found = 0

while n_items_found > prev_items_found:
    prev_items_found = n_items_found
    drill_cards = browser.find_elements_by_css_selector(sel)
    n_items_found = len(drill_cards)
    print(n_items_found)
    last_drill_card = drill_cards[-1] 
    loc = last_drill_card.location_once_scrolled_into_view
    time.sleep(2)
```

```python
card = drill_cards[0]
```

```python
def get_card_price(card):
    """Returns the price from a Sears product card."""
    sel = "span.card-price"
    try:
        price_text = card.find_element_by_css_selector(sel).text
        price = float(price_text.lstrip('$'))
        return price
    except NoSuchElementException:
        return None
    
def get_card_title(card):
    """Returns the title from a Sears product card."""
    sel = "h3.card-title a"
    title = card.find_element_by_css_selector(sel).text
    return title

def get_card_stars(card):
    """Returns the star rating from a Sears product card."""
    sel = 'span.stars'
    star_span = card.find_element_by_css_selector(sel)
    classes = star_span.get_attribute('class').split()
    stars = classes[-1]
    if stars == 's0':
        return None
    stars = stars.lstrip('s')
    stars = float(stars) / 10
    return stars
```

```python
def parse_card(card):
    return {
        'title': get_card_title(card),
        'price': get_card_price(card),
        'stars': get_card_stars(card),
    }
```

```python
data = [parse_card(card) for card in drill_cards]
```
