```python
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver import Firefox
from string import ascii_lowercase as lower
from string import ascii_uppercase as upper
import random
import time
```

```python
def click_close_links(browser):
    """Finds the link to close the popup box."""
    sel = ".bx-close-link"
    close_links = browser.find_elements_by_css_selector(sel)
    for i, link in enumerate(close_links):
        print(f"{i}: ");
        try:
            link.click()
            print("success!")
        except Exception as e:
            print(e)
```

```python
def click_search(browser):
    """Clicks on the search box."""
    sel = "div.KeySearch input#keyword"
    box = browser.find_element_by_css_selector(sel)
    box.click()
    return box

```

```python

def type_like_a_human(box,
                      target_item,
                      delay=.2,
                      p_typo=.2):
    """Type realistically into a search box."""
    for key in target_item:
        if random.random() < p_typo:
            box.send_keys(random.choice(lower))
            time.sleep(delay*2)
            box.send_keys('\b')
        time.sleep(random.random()*delay/2 + delay/2)
        box.send_keys(key)
        
```

```python
def get_all_item_cards(browser):
    """Scrolls product cards into the window."""
    sel = 'div.card-container'
    prev_items_found = -1
    n_items_found = 0
    while n_items_found > prev_items_found:
        prev_items_found = n_items_found
        drill_cards = browser.find_elements_by_css_selector(sel)
        n_items_found = len(drill_cards)
        print(n_items_found)
        last_drill_card = drill_cards[-1] 
        loc = last_drill_card.location_once_scrolled_into_view
        time.sleep(2)
    return drill_cards
```

```python
def get_card_price(card):
    """Returns the price from a Sears product card."""
    sel = "span.card-price"
    try:
        price_text = card.find_element_by_css_selector(sel).text
        price = float(price_text.lstrip('$'))
        return price
    except NoSuchElementException:
        return None
    
def get_card_title(card):
    """Returns the title from a Sears product card."""
    sel = "h3.card-title a"
    title = card.find_element_by_css_selector(sel).text
    return title

def get_card_stars(card):
    """Returns the star rating from a Sears product card."""
    sel = 'span.stars'
    star_span = card.find_element_by_css_selector(sel)
    classes = star_span.get_attribute('class').split()
    stars = classes[-1]
    if stars == 's0':
        return None
    stars = stars.lstrip('s')
    stars = float(stars) / 10
    return stars
```

```python
def parse_card(card):
    return {
        'title': get_card_title(card),
        'price': get_card_price(card),
        'stars': get_card_stars(card),
    }
```

```python
data = [parse_card(card) for card in cards]
```
